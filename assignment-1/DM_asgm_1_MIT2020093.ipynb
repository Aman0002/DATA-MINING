## **Reading the Data**

```
import pandas as pd
import math as math
import numpy as np
#Reading the csv file
df = pd.read_csv('/content/crop_production.csv - crop_production.csv.csv')
crop_year = set(df["Crop_Year"])
df
```

## **Transform crop production dataset into following format**

```

#Transformed variable our crop-production dataset in the required format
transformed = df.pivot_table(index = ["State_Name","District_Name","Season","Crop"],values=["Area","Production"],columns = "Crop_Year")

transformed = transformed.swaplevel(0,1,axis=1).sort_index(axis=1)

transformed
```

## **Slice the transform dataset in two different dataframe**
* Df1= data[:50%]
* Df2=data[51%:]


```
row , col = transformed.shape
df1 = transformed[:int(row/2)]
df2 = transformed[int(row/2):]
print("DATA FRAME 1 ")
print(df1.head())

print("DATA FRAME 2")
print(df2.head())
```

## **Find the count of NaN values attribute(feature) wise in both dataframes**

```
print("NaN values in the Data Frame 1 are :-\n")
print(df1.isna().sum())
print()
print("NaN values in the Data Frame 2 are :-\n")
print()
print(df2.isna().sum())

```

## **Use â€˜ffillâ€™ to replace NaN values in Df1**


```
print("NaN values in the Data Frame 1 before forward filling :-\n")
print(df1.isna().sum())
df1 = df1.fillna(method='ffill')
print("NaN values in the Data Frame 1 after forward filling :-\n")
print(df1.isna().sum())
```

## **Use â€˜bfillâ€™ to replace NaN values in Df2**


```
print("NaN values in the Data Frame 2 before backward filling :-\n")
print(df2.isna().sum())
df2 = df2.fillna(method='bfill')
print("NaN values in the Data Frame 2 after backward filling :-\n")
print(df2.isna().sum())
```

## **Find the mean value of every attribute and replace with NaN in df1**

```
keys = list(df1.keys())
Area_mean={}
Production_mean={}
for i in keys:
  m_area = df1[i[0]]["Area"].mean()
  m_production = df1[i[0]]["Production"].mean()
  Area_mean[i[0]]={"Area":m_area}
  Production_mean[i[0]]={"Production":m_production}
  df1[i[0]]["Area"].fillna(m_area,inplace = True)
  df1[i[0]]["Production"].fillna(m_production,inplace = True)

print("mean of Area :" ,Area_mean)
print()
print("mean of Production :" ,Production_mean)
print()
df1.head(50)
```

## **Find the median value of every attribute and replace with NaN in df2**


```
keys = list(df2.keys())
Area_median={}
Production_median={}
for i in keys:
  med_area = df2[i[0]]["Area"].median()
  med_production = df2[i[0]]["Production"].median()
  Area_median[i[0]]={"Area":med_area}
  Production_median[i[0]]={"Production":med_production}
  df2[i[0]]["Area"].fillna(med_area,inplace = True)
  df2[i[0]]["Production"].fillna(med_production,inplace = True)
print("median of Area :" ,Area_median)
print()
print("median of Production :" ,Production_median)
print()
df2.head(50)
```

## **Convert df2 into standardization format**

```
from sklearn.preprocessing import StandardScaler

standardscaler = StandardScaler()

for year in crop_year:
  tempi = pd.DataFrame(df2[year], columns =["Area","Production"], )
  tempi = standardscaler.fit_transform(tempi)
  df2[year] = standardscaler.transform(tempi)


df2.head(50)
```
