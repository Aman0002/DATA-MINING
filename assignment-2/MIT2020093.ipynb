## Prepare the Data

```
import pandas as pd
import itertools

min_support_count =  40
data = pd.read_csv("/content/store_data.csv - store_data.csv.csv",header= None)

records = [] #list 
row,col = data.shape

for i in range(row):
  records.append([str(data.values[i,j]) for j in range(col)])

items = [] #list
for sublist in records:
  for item in sublist:
    if (item!='nan'):
      items.append(item)

items.sort()
items
```

## STAGE-1 Taking One Item at a Time

```
 
 
c1 = {i: items.count(i) for i in items}

l1 = {}
print(c1)
for key,value in c1.items():
  if value  >= min_support_count:
    l1[key] = value 

print(l1)
len(l1)
```

## STAGE-2 Taking two items at a time

```
L1 = list(itertools.combinations(l1,2))
c2 = {}
l2 = {}

for i in L1:
  count = 0
  for j in records:
    if(set(i) <= set(j)):
      count+= 1
  c2[i]= count

def check_subset_frequency(itemset, l, n):
    if n>1:    
        subsets = list(itertools.combinations(itemset, n))
    else:
        subsets = itemset
    for iter1 in subsets:
        if not iter1 in l:
            return False
    return True

for key,value in c2.items():
  if (value>min_support_count):
    if(check_subset_frequency(key,l1,1)):
      l2[key] = value


print(len(l2))

l2 = list(l2.keys())
```

## Stage-3 taking three items at a time

```

L2 = sorted(list(set([item for t in l2 for item in t])))
L2 = list(itertools.combinations(L2, 3))
print(len(L2))
c3 = {}
l3 = {}
for iter1 in L2:
  count = 0
  for iter2 in records:
    if(set(iter1) <= set(iter2)):
      count+= 1
  c3[iter1] = count

for key, value in c3.items():
  if value >= min_support_count:
    if check_subset_frequency(key, l2, 2):
      l3[key] = value 

l3 = list(l3.keys())#making dict into lists
```

## Making Association Rules

```

sets = []
for iter1 in l3:
    subsets = list(itertools.combinations(iter1, 2))
    sets.append(subsets)
#print(sets)
def support_count(itemset, itemlist):
    return itemlist[itemset]

for i in range(0, len(l3)):
    for iter1 in sets[i]:
        a = iter1
        b = set(l3[i]) - set(iter1)
        confidence = (support_count(l3[i], c3)/support_count(iter1, c2))*100
        print("{}    ->       {}  confidence = {}% {}".format(a, b, confidence, "STRONG" if (confidence >= 70) else "WEAK"))
```
